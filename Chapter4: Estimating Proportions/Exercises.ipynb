```python
import numpy as np
from empiricaldist import Pmf
import matplotlib.pyplot as plt
from scipy.stats import binom  
```

# Exercise 4-1

```python
# Define the data
n = 3  # Number of attempts
x = 3  # Number of hits


# Define a range of possible values for p
p_values = np.linspace(0, 1, 101)  # Values between 0 and 1 with 0.01 increments

prior = Pmf(1,p_values,name="Uniform")

# Calculate the likelihood for each value of p
likelihood = binom.pmf(x, n, p_values)

posterior = prior * likelihood
posterior.normalize()

# Plot the posterior distribution
plt.plot(p_values, posterior, label='Posterior Probability')
plt.xlabel('Hitting Probability (p)')
plt.ylabel('Posterior Probability')
plt.legend()
plt.show()
```
